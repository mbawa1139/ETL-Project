```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Import Accidents csv
csv_file = "Accidents0515.csv"
accidents_df = pd.read_csv(csv_file)
accidents_df.head()
```

```python
# Import Casualties csv
csv_file = "Casualties0515.csv"
casualties_df = pd.read_csv(csv_file, error_bad_lines=False)
casualties_df.head()
```

```python
# Import Casualties csv
csv_file = "Vehicles0515.csv"
vehicles_df = pd.read_csv(csv_file, error_bad_lines=False)
vehicles_df.head()
```

```python
# Narrow down the Accidents data frame to only accidents occurring after 2010

after_2010_accidents = accidents_df[accidents_df['Accident_Index'].str.startswith('201')]
after_2010_accidents
```

```python
# Narrow down the Casualties data frame to only accidents occurring after 2010

after_2010_casualties = casualties_df[casualties_df['Accident_Index'].str.startswith('201')]
after_2010_casualties
```

```python
# Narrow down the Vehicles data frame to only accidents occurring after 2010

after_2010_vehicles = vehicles_df[vehicles_df['Accident_Index'].str.startswith('201')]
after_2010_vehicles
```

```python
# Narrow down columns to only the index and location

new_accidents_df = after_2010_accidents[["Accident_Index", "Longitude", "Latitude"]]
new_accidents_df
```

```python
# Narrow down columns to only the index, sex and age of casualties

new_casualties_df = after_2010_casualties[["Accident_Index", "Sex_of_Casualty", "Age_of_Casualty"]]
new_casualties_df
```

```python
# Narrow down columns to only the index and vehicle types

new_vehicles_df = after_2010_vehicles[["Accident_Index", "Vehicle_Type"]]
new_vehicles_df
```

```python
# Merge accidents and vehicles on accident index into a new df

accidents_vehicles = new_vehicles_df.merge(new_accidents_df, how = "left", on="Accident_Index")
accidents_vehicles
```

```python
# Merge casualty df into the previous merged df and save as new df

a_v_c_df = accidents_vehicles.merge(new_casualties_df, how = "left", on="Accident_Index")
a_v_c_df
```

```python
# Use a dictionary to go through the 'Sex_of_Casualty' column and replace the number
# value with the actual gender
dict = {1: "Male", 2: "Female", 3: "Not Known", 4: "Data missing or out of range"}

a_v_c_df["Sex_of_Casualty"] = a_v_c_df["Sex_of_Casualty"].map(dict)
```

```python
# Use a dictionary to go through the 'Vehicle_Type' column and replace the motorcycle indexes
# to actual 'Motorcycle' and leave all other vehicle types alone

dict = {2: "Motorcycle", 3: "Motorcycle", 4: "Motorcycle", 5: "Motorcycle"}
a_v_c_df["Vehicle_Type"] = a_v_c_df["Vehicle_Type"].map(dict)
```

```python
# Drop all rows with NA
motorcycle_accident_df = a_v_c_df.dropna()
```

```python
motorcycle_accident_df
```

```python
sql_lite = "sqlite:///etl_sqlite.db"

# Initial DB connection
engine = create_engine(sql_lite)
```

```python
# Read DF into DB
motorcycle_accident_df.to_sql(name='motorcycle_accident', con=engine, if_exists='append', index=False)
```

```python
# Check connection of DB is working
engine.table_names()
```

```python
pd.read_sql_query('Select * from motorcycle_accident', con=engine).head()
```

```python
# Make html out of DB table
motorcycle_accident_html = pd.read_sql_query('Select * from motorcycle_accident', con=engine).to_html()
```

```python
# Write html to a file
write_file = open("motorcycle_accident.html", "w")
write_file.write(motorcycle_accident_html)
write_file.close()
```
